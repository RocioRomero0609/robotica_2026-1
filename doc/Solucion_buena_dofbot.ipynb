## Planteamiento de la solución de la cinemática inversa de postura del robot

El planteamiento de la cinemática inversa se establece en función de las restricciones cinemáticas de la postura del robot, ya que el sistema se considera subactuado.


$${}^O{\xi }_P=\left\lbrack \begin{array}{c} {}^Ox_P\newline {}^Oy_P\newline {}^Oz_P\newline {}^O{\gamma }_P\newline {}^O{\beta }_P\newline {}^O{\alpha }_P \end{array}\right\rbrack$$


Donde la posición del sistema de referencia asociado al efector final son restricciones que el robot puede cumplir.


$${}^O{\mathbf{p}}_P=\left\lbrack \begin{array}{c} {}^Ox_P\newline {}^Oy_P\newline {}^Oz_P \end{array}\right\rbrack$$





En el caso de las restricciones de orientación del gripper la orientación con respecto al eje ${\hat{\mathbf{z}} }_O$ que corresponde al ángulo ${}^O{\alpha }_P$, este corresponde con el ángulo de la cintura del robot ${}^O{\theta }_1$ el cual corresponde al plano de simetria del robot cuya orientación:


$${}^O{\theta }_1=\arctan \left(\frac{{}^Oy_P}{{}^Ox_P}\right)$$





Las restricciones de orientación del gripper con respecto a los ejes ${\hat{x} }_O$ y ${\hat{y} }_O$ no pueden ser cumplidas debido a las restricciones mecánicas que la configuración del robot, estás restricciones solamente pueden ser satisfechas sobre la orientación del sistema {1}. Con base en lo anterior, para simplificar el calculo de la solución esta se planteará en el sistema {1}.






Sobre el plano xz del sistema {1} el problema se pue solucionar mediante el análisis geometrico de la cadena cinemática del robot. Para determinar la solición, es necesario expresar el vector de posición del sistema {P} con respecto al sistema {1}.


$${}^1{\mathbf{p}}_P={}^1{\mathbf{R}}_O\left({}^O{\mathbf{p}}_P-{}^O{\mathbf{p}}_1\right)$$


Donde la matriz ${}^1{\mathbf{R}}_O$ es una matriz de rotación con respecto al eje ${\hat{\mathbf{z}} }_O$ el ángulo ${}^O{\theta }_1$, el vector ${}^1{\mathbf{p}}_P$ esta alineado sobre el plano de simetria del robot, por lo que las componentes de este vector en función de la posición del sistema {P} con respecto al sistema {O} es:


$${}^1{\mathbf{p}}_P=\left\lbrack \begin{array}{c} \sqrt{{{}^Ox_P}^2 +{{}^Oy_P}^2 }\newline 0\newline {}^Oz_P-\left\|{}^O{\mathbf{p}}_1\right\| \end{array}\right\rbrack$$


Donde $\left\|{}^O{\mathbf{p}}_1\right\|$ corresponde al parámetro ${}^Oz_1$, que corresponde a la altura de la articulación del hombro con respecto al plano xy del sistema {O}.


Ya cálculada la posición del sistema {P} con respecto al sistema {1}, es necesario calcular la posición del sistema {4} en términos del punto ${}^1{\mathbf{p}}_P$ y el ángulo ${}^1{\theta }_P$.


$${}^4{\mathbf{p}}_P=\left\lbrack \begin{array}{c} \sqrt{{{}^Ox_P}^2 +{{}^Oy_P}^2 }-L_3 \;\sin \left({}^1{\theta }_P\right)\newline 0\newline {}^Oz_P-\left\|{}^O{\mathbf{p}}_1\right\|-L_3 \;\cos \left({}^1{\theta }_P\right) \end{array}\right\rbrack$$

```matlab
% Parametros del robot
z_O1 = 0.105;
L1 = 0.084;
L2 = 0.084;
L3 = 0.115;

% Requerimientos de posición
x_OP = 0.2;
y_OP = 0.0;
z_OP = 0.05;

theta_1P = 8*pi/16;
% Calculo de la posición %%%%%

p_OP = [x_OP; y_OP; z_OP];

% Primer angulo del robot
theta_O1 = atan2(y_OP,x_OP)

% Calculo de la posición del sistema P con respecto al sistema 1

p_1P = [sqrt(x_OP^2+y_OP^2); 0;z_O1 - z_OP];

p_4P = p_1P - [L3*sin(theta_1P); 0; L3*cos(theta_1P)];

Norm_p_4P = norm(p_4P)

epsilon = acos(p_4P(3)/Norm_p_4P )

alpha = acos((L1^2 + Norm_p_4P^2 - L2^2) / (2*L1*Norm_p_4P))

theta_12 = epsilon - alpha

beta = asin((sin(alpha) * Norm_p_4P) / (L2));

theta_23 = pi - beta

theta_34 = theta_1P - theta_12 - theta_23
```
